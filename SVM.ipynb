```python
import pandas as pd
```

```python
df=pd.read_csv("Automobile_data_cleansed.csv")
```

```python
df.head()
```

```python
df['luxary']=[1 if x>20000 else 0 for x in df['price']]
```

```python
df.head()
```

```python
df=df.drop('price',axis=1)
```

```python
df=df.drop('Unnamed: 0' ,axis=1)
```

```python
x=df.drop('luxary',axis=1)
y=df['luxary']
```

```python
df.head()
```

```python
x=pd.get_dummies(x)
```

```python
x.shape
```

```python
import sklearn.model_selection
x_train,x_test,y_train,y_test=sklearn.model_selection.train_test_split(x,y,test_size=0.30,random_state=5)
```

```python
from sklearn import svm
```

```python
svmclassifier=svm.SVC(kernel='poly')
```

```python
svmclassifier.fit(x_train,y_train)
```

```python
y_pred_poly_2=svmclassifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_poly_2)
```
