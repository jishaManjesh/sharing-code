```python
movies=["gaya" 1999,"hai"1998,"ghee"2000,"lime"1990,"ghaa"2003,"jahh"2000,"gthh"2009]
gmovies=[]
for titile in movies:
    if titile.startswith("g"):
        gmovies.append(titile)
        print(gmovies)
```

```python
gmovies=[titile for titile in movies if titile.startswith("g")]
print(gmovies)
```

```python
gmovies=[hed for hed in movies if hed.startswith("g")]
print(gmovies)
```

```python
yearm = [titile for (titile,year) in movies if year < 2000]
print(yearm)
```
