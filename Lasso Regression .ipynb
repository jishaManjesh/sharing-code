```python
data_dummy=pd.get_dummies(data)
```

```python
data_dummy.columns
```

```python
dellist=['Marital_divorced','Education_primary','Communication_telephone''LastContactMonth_apr','Outcome','job_other','OutCome_No Response']
```

```python
data_dummy=data_dummy.drop(dellist,axis=1)
```

```python
data_dummy.head()
```

```python
X=data_dummy.drop('CarInsurance',axis=1)
y=data['CarInsurance']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=1/3,random_state=42)
```

```python
## MODEL
```

```python
# Define Model parameters to tune
model_parameters = {
        'penalty':['l1'],
        'C': [1,10,100,1000],
        'class_weight': ['balanced', None]
    }
```

```python
from sklearn.linear_model import LogisticRegression
?LogisticRegression
```

```python
from sklearn.model_selection import GridSearchCV

model = LogisticRegression(random_state=1)
gscv = GridSearchCV(estimator=model, 
                    param_grid=model_parameters, 
                    scoring='roc_auc'
                    )

gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
# Re-fit the model with the best parameters
final_mod = LogisticRegression(**gscv.best_params_)
final_mod.fit(xtrain, ytrain)
```

```python
final_mod.coef_[0]
```

```python
y_pred=final_mod.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,y_pred))
```

```python
simp_mod = LogisticRegression()
simp_mod.fit(xtrain,ytrain)
y_simp=simp_mod.predict(xtest)
```
