```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn
import statsmodels.formula.api as sm
```

```python
df=pd.read_csv("E:/dataScience/linear regression/Automobile_data.csv")
```

```python
df.head().T
```

```python
normlossmean=pd.to_numeric(df[df['normalized-losses']!='?']['normalized-losses']).mean()
normlossmean
```

```python
df['normalized-losses']=pd.to_numeric(df['normalized-losses'].replace('?',normlossmean))
```

```python
nodoormean=(df[df['num-of-doors']!='?']['num-of-doors']).mode()[0]
df['num-of-doors']=df['num-of-doors'].replace('?',nodoormean)
```

```python
bore_mean=pd.to_numeric(df[df['bore']!='?']['bore']).mean()
df['bore']=pd.to_numeric(df['bore'].replace('?',bore_mean))
```

```python
stroke_mean=pd.to_numeric(df[df['stroke']!='?']['stroke']).mean()
df['stroke']=pd.to_numeric(df['stroke'].replace('?',stroke_mean))
```

```python
hp_mean=pd.to_numeric(df[df['horsepower']!='?']['horsepower']).mean()
df['horsepower']=pd.to_numeric(df['horsepower'].replace('?',hp_mean))
```

```python
peak_rpmmean=pd.to_numeric(df[df['peak-rpm']!='?']['peak-rpm']).mean()
df['peak-rpm']=pd.to_numeric(df['peak-rpm'].replace('?',peak_rpmmean))
```

```python
price_mean=pd.to_numeric(df[df['price']!='?']['price']).mean()
df['price']=pd.to_numeric(df['price'].replace('?',price_mean))
```

```python
df1=df
```

```python
df1=pd.get_dummies(df1)
df1.T
```

```python
dep='price'
X=df1.drop(dep, axis=1)
Y=df1[dep]
print(X)
print(Y)
```

```python
import sklearn.model_selection
X_train,X_test,Y_train,Y_test=sklearn.model_selection.train_test_split(X,Y,test_size=0.30,random_state=5)
```

```python
from sklearn.linear_model import LinearRegression
reg =LinearRegression()
reg.fit(X_train,Y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
import math
mean_squared_error(y_pred,Y_test)
```

```python
math.sqrt(sum((y_pred-Y_test)**2)/len(y_pred))
```
